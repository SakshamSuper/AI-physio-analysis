```python

# Initialize MediaPipe
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Calculate angle between 3 points
def calculate_angle(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    angle = math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
    return abs((angle + 360) % 360)

# Get pixel coordinates from landmark index
def get_coords(landmarks, idx, width, height):
    lm = landmarks[idx]
    return int(lm.x * width), int(lm.y * height)

cap = cv2.VideoCapture(0)

with mp_pose.Pose(min_detection_confidence=0.6, min_tracking_confidence=0.6) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)
        h, w, _ = frame.shape
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(image_rgb)
        image = frame.copy()

        if results.pose_landmarks:
            lm = results.pose_landmarks.landmark

            # Get keypoints
            nose = get_coords(lm, mp_pose.PoseLandmark.NOSE.value, w, h)
            l_shldr = get_coords(lm, mp_pose.PoseLandmark.LEFT_SHOULDER.value, w, h)
            r_shldr = get_coords(lm, mp_pose.PoseLandmark.RIGHT_SHOULDER.value, w, h)
            l_hip = get_coords(lm, mp_pose.PoseLandmark.LEFT_HIP.value, w, h)
            r_hip = get_coords(lm, mp_pose.PoseLandmark.RIGHT_HIP.value, w, h)
            l_knee = get_coords(lm, mp_pose.PoseLandmark.LEFT_KNEE.value, w, h)
            r_knee = get_coords(lm, mp_pose.PoseLandmark.RIGHT_KNEE.value, w, h)

            mid_shldr = ((l_shldr[0] + r_shldr[0]) // 2, (l_shldr[1] + r_shldr[1]) // 2)
            mid_hip = ((l_hip[0] + r_hip[0]) // 2, (l_hip[1] + r_hip[1]) // 2)
            vertical_up = (mid_shldr[0], mid_shldr[1] - 100)

            avg_knee_y = (l_knee[1] + r_knee[1]) // 2
            sitting = abs(mid_hip[1] - avg_knee_y) < 50

            # Neck Tilt Angle
            neck_angle = calculate_angle(nose, mid_shldr, vertical_up)
            neck_color = (0, 255, 0) if neck_angle < 10 else (0, 0, 255)
            cv2.line(image, nose, mid_shldr, neck_color, 2)
            cv2.putText(image, f"Neck Tilt: {neck_angle:.1f}°", (30, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, neck_color, 2)
            if neck_angle >= 10:
                cv2.putText(image, "⚠️ Keep your head upright!", (30, 90),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

            # Leaning Forward Proxy using Nose–Shoulder–Hip Angle
            lean_angle = calculate_angle(nose, mid_shldr, mid_hip)
            lean_color = (0, 255, 0) if lean_angle > 155 else (0, 0, 255)
            cv2.line(image, nose, mid_shldr, lean_color, 2)
            cv2.line(image, mid_shldr, mid_hip, lean_color, 2)
            cv2.putText(image, f"Lean Angle: {lean_angle:.1f}°", (30, 120),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, lean_color, 2)
            if lean_angle < 150:
                cv2.putText(image, "⚠️ Sit straight, avoid leaning forward!", (30, 150),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

            # Spine Symmetry
            shoulder_diff = abs(l_shldr[1] - r_shldr[1])
            spine_color = (0, 255, 0) if shoulder_diff < 15 else (0, 0, 255)
            cv2.line(image, l_shldr, r_shldr, spine_color, 2)
            cv2.line(image, l_hip, r_hip, spine_color, 2)
            cv2.putText(image, f"Spine Shift: {shoulder_diff}px", (30, 180),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, spine_color, 2)
            if shoulder_diff >= 15:
                cv2.putText(image, "⚠️ Balance your shoulders evenly!", (30, 210),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

            # Posture Summary
            posture_good = (neck_angle < 10 and lean_angle > 155 and shoulder_diff < 15)
            posture_status = "✅ Good Posture" if posture_good else "❌ Needs Correction"
            posture_color = (0, 255, 0) if posture_good else (0, 0, 255)
            cv2.putText(image, f"Posture: {posture_status}", (30, 250),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, posture_color, 2)

            # Sitting or Not
            sit_status = "🪑 Sitting" if sitting else "⚠️ Not Sitting"
            cv2.putText(image, f"Detected: {sit_status}", (30, 280),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2)

        cv2.imshow("Front View Sitting Posture Monitor", image)
        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
